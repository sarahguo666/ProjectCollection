# Probability density function of exponential distribution 

Note that the probability density function of exponential distribution with parameter $\lambda$ looks like this

$\begin{equation}
    f(x)=
    \begin{cases}
       \lambda e^{-\lambda x}, & \text{if}\ x>0 \\
      0, & \text{otherwise}
    \end{cases}
  \end{equation}$




```python
import numpy as np
import matplotlib.pyplot as plt

pdf_exp = lambda x, lam: lam*np.exp(-lam*x)*(x>0)

def pdf_exp2(x,lam):
    return lam*np.exp(-lam*x)*(x>0)

x = np.array([1,2,3]) 
lam = 1
print(pdf_exp(x,lam))
print(pdf_exp2(x,lam))
```

```python
xpts=np.arange(-2,10,0.05)

y = pdf_exp(xpts, 2)

plt.figure(figsize=(8, 5))

plt.plot(xpts,y,'.')

plt.xlabel("x")
plt.ylabel("exponential with parameter 2 pdf ")

plt.show()
```

# For different values of $\lambda$ plot probability density function on the same figure

```python
xpts=np.arange(1e-10,10,0.05)

plt.figure(figsize=(12, 9))

plt.style.use('ggplot')

for l in np.arange(0.5,5,0.5):#[0.5,1,2]:#
    y = pdf_exp(xpts, l)
    plt.plot(xpts,y,'-',label = "$\lambda = %.1f$"%l)
    plt.xlabel("x")
    plt.ylabel("exponential pdf ")
    
plt.legend()
plt.title("pdf of an Exponential distribution varying parameter $\lambda$");
plt.show()
```
